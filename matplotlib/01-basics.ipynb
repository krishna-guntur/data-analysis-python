Basics of Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,10)
```

```python
y = 3*x 
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axix')
plt.title('y = mx')
plt.xlim(-8, 8)
plt.ylim(-16,16)

plt.savefig(r'.\visualizations\y=mx.png')
```
