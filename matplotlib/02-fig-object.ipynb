```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**2
```

```python
x = np.arange(0,10)
y = x**3
```

```python
fig = plt.figure(dpi=100, figsize=(5,3))

axes = fig.add_axes([0,0,1,1])
#axes2 = fig.add_axes([0.5,0.5,0.5,1])

axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_title('y = x²')
axes.set_xlim(0,10)
axes.set_ylim(0,125)

axes.plot(a,b)
axes.plot(x,y)

fig.savefig(r'.\visualizations\y=mx²', bbox_inches='tight' )
```
