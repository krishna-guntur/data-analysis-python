```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
a = np.linspace(0,10,11)
b = a**2
```

```python
x = np.arange(0,10)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a,b)
axes[1].plot(x,y)

axes[0].set_title('y=mxÂ²')
axes[1].set_title('y=sinx')

for axis in axes:
  axis.set_xlabel('x-axis')
  axis.set_ylabel('y-axis')

fig.suptitle('Subplots')

plt.tight_layout()

fig.savefig(r'.\visualizations\subplots.png', bbox_inches='tight')
```
