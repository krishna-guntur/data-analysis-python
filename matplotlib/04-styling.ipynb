```python
import numpy as np, matplotlib.pyplot as plt
```

```python
x = np.linspace(0,25,25)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='x, x')
ax.plot(x,y, label='y=tan(x)')

ax.legend(loc=0)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='x, x')
ax.plot(x,y, label='y=tan(x)')

ax.legend(loc=(1.01,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='x, x')
ax.plot(x,y, label='y=tan(x)')

ax.legend(loc=(0.5,-0.12))
```

Using Custom Colors

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='x, x', color='#008080')
ax.plot(x,y, label='y=tan(x)', color='#8F00FF')

ax.legend(loc=(1.01,0.5))
```

```python
fig = plt.figure(dpi=500)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='x, x', color='#008080', lw=2)
ax.plot(x,y, label='y=x²', color='#8F00FF', lw=3)

ax.set_xlim(0,27)
ax.set_ylim(0,700)

ax.legend(loc=(1.01,0.5))

fig.savefig(r'.\visualizations\styling.png', bbox_inches='tight')
```
