```python
def last_four_digits(num):
  return str(num)[-4:]

def price_category(price):
  if price < 13.34:
    return '$'
  elif price > 13.34 and price < 24.12:
    return '$$'
  else:
    return '$$$'
```

```python
df['CC Number_4'] = df['CC Number'].apply(last_four_digits)

df['Price Category'] = df['total_bill'].apply(price_category)

df.head(10)
```
