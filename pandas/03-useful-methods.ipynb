```python
df.describe()
```

```python
def last_four_digits(num):
  return str(num)[-4:]

def price_category(price):
  if price < 13.34:
    return '$'
  elif price > 13.34 and price < 24.12:
    return '$$'
  else:
    return '$$$'
```

```python
df['CC Number_4'] = df['CC Number'].apply(last_four_digits)

df['Price Category'] = df['total_bill'].apply(price_category)

df.head(10)
```

```python
df['total_bill'].apply(lambda bill: bill*10)
```

Use Multiple Columns

```python
def quality(total_bill, tip):
  if tip/total_bill > 0.18:
    return 'Generous'
  else:
    return 'Cheap'
```

```python
df['tip_quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'], df['tip']),axis=1)
```

```python
df[df['tip_quality'] == 'Generous']
```

```python
np.vectorize(quality)(df['total_bill'],df['tip'])
```
