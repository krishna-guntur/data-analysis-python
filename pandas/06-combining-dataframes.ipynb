Merge is similar to join in SQL

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, how='inner', on='name')
```

```python
pd.merge(registrations, logins, how='outer', on='name')
```

```python
pd.merge(registrations, logins, how='left', on='name')
```

```python
pd.merge(registrations, logins, how='right', on='name')
```

```python
registrations.columns = ['reg_id', 'reg_name']
```

```python
registrations
```

```python
pd.merge(left=registrations, right=logins, how='outer', left_on='reg_name', right_on='name')
```

```python
pd.merge(left=registrations, right=logins, how='inner', left_on='reg_name', right_on='name')
```

```python
pd.merge(left=registrations, right=logins, how='left', left_on='reg_name', right_on='name')
```

```python
pd.merge(left=registrations, right=logins, how='right', left_on='reg_name', right_on='name')
```
