```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\codebase\data-analysis-python\datasets\StudentsPerformance.csv')

df.head()
```

```python
plt.figure(figsize=(10, 5), dpi=500)

sns.jointplot(data=df, x='math score', y='reading score'
              ,kind='hex')

plt.savefig(r'.\visualizations\joint-plot-hex.png')
```

```python
plt.figure(figsize=(10, 5), dpi=500)

sns.jointplot(data=df, x='math score', y='reading score'
              ,kind='scatter')

plt.savefig(r'.\visualizations\joint-plot-scatter.png')
```

```python
plt.figure(figsize=(10, 5), dpi=500)

sns.jointplot(data=df, x='math score', y='reading score'
              ,kind='scatter', hue='gender')

plt.savefig(r'.\visualizations\joint-plot-scatter-hue.png')
```

Pair Plot

```python
sns.pairplot(data=df, hue='gender')

plt.savefig(r'.\visualizations\pair-plot.png')
```

```python
sns.pairplot(data=df, hue='gender', corner=True)

plt.savefig(r'.\visualizations\pair-plot-corner.png')
```
