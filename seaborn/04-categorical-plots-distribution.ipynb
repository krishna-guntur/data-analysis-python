```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\codebase\data-analysis-python\datasets\StudentsPerformance.csv')

df.head()
```

```python
df
```

Box-Whisker Plot

```python
plt.figure(figsize=(10, 5), dpi=500)
sns.boxplot(data=df, x='math score', y='test preparation course'
            ,hue='lunch')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\box-whisker-plot.png')
```

Violin Plot

```python
plt.figure(figsize=(10, 5), dpi=500)
sns.violinplot(data=df, x='math score', y='test preparation course'
            ,hue='lunch')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\violin-plot.png')
```

Using split

```python
plt.figure(figsize=(10, 5), dpi=500)
sns.violinplot(data=df, x='math score', y='test preparation course'
            ,hue='lunch', split=True, inner='quartile')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\violin-plot-split.png')
```

Swarm Plot

```python
plt.figure(figsize=(10, 5), dpi=500)
sns.swarmplot(data=df, x='math score', y='gender'
            ,size=5, hue='test preparation course')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\swarm-plot.png')
```

Boxen Plot

```python
plt.figure(figsize=(10, 5), dpi=500)
sns.boxenplot(data=df, x='math score', y='gender'
            , hue='test preparation course')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\boxen-plot.png')
```
