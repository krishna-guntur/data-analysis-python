```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\codebase\data-analysis-python\datasets\StudentsPerformance.csv')

df.head()
```

```python
plt.figure(figsize=(10,5), dpi=500)

sns.catplot(data=df, x='reading score', y='gender'
            ,hue='lunch', kind='box'
            )

plt.savefig(r'.\visualizations\cat-plot.png')
```

```python
plt.figure(figsize=(10,5), dpi=500)

sns.catplot(data=df, x='reading score', y='gender'
            ,hue='lunch', kind='boxen'
            ,row='test preparation course')

plt.savefig(r'.\visualizations\cat-plot-row.png')
```

Pair Grid

```python
g = sns.PairGrid(data=df, hue='gender')

g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.scatterplot)
g = g.map_diag(sns.histplot)

plt.savefig(r'.\visualizations\pair-grid.png')
```
