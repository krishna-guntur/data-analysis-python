```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r'D:\codebase\data-analysis-python\datasets\dm_office_sales.csv')

df.head()
```

```python
plt.figure(figsize=(10,4), dpi = 500)
plt.ylim(0, 150)

sns.countplot(data=df, x='division', hue='level of education')

plt.savefig(r'.\visualizations\count-plot.png')
```

Bar Plot

```python
plt.figure(figsize=(10,4), dpi = 500)

sns.barplot(data=df, x='level of education', y='salary', hue='division')

plt.legend(bbox_to_anchor=(1,0.75))

plt.savefig(r'.\visualizations\bar-plot.png')
```
