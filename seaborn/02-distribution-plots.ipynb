```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

Rug Plot

```python
df = pd.read_csv(r'D:\codebase\data-analysis-python\datasets\dm_office_sales.csv')
```

```python
df
```

```python
sns.rugplot(x='salary', data=df, height=0.2)

plt.savefig(r'.\visualizations\rug-plot.png')
```

DisPlot

```python
sns.displot(x='salary', data=df, bins=20)

plt.savefig(r'.\visualizations\dis-plot.png')
```

HistPlot

```python
sns.histplot(data=df, x='salary')

plt.savefig(r'.\visualizations\hist-plot.png')
```

```python
sns.histplot(data=df, x='salary', kde=True)

plt.savefig(r'.\visualizations\hist-plot-kde.png')
```

```python
np.random.seed(42)

age_samples = pd.DataFrame(np.random.randint(0,100,200),columns=['age'])

age_samples
```

```python
sns.displot(data=age_samples, x='age', bins=30)
```

```python
sns.kdeplot(data=age_samples, x='age')

plt.savefig(r'.\visualizations\kde-plot.png')
```

```python
sns.kdeplot(data=age_samples, x='age', clip=[0,100])
```
